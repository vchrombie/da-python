```python
def signature(word):
    return ''.join(sorted(word))
```

```python
signature("winemine")
```

```python
def anagrams(mywords):
    return [word for word in wordclean if signature(word) == signature(mywords)]
```

```python
anagrams('cat')
```

```python
%timeit anagrams('cat')
```

```python
words_bysig = {}
for word in wordclean:
    words_bysig[signature(word)].append(word)
```

```python
import collections
words_bysig = collections.defaultdict(list)
for word in wordclean:
    words_bysig[signature(word)].append(word)
words_bysig
```

```python
def anagram_fast(mywors):
    return words_bysig[signature(mywors)]
```

```python
%timeit anagram_fast('cat')
```

```python
wordlen = collections.defaultdict(list)

for word in wordclean:
    wordlen[len(word)].append(word)
wordlen

    
```

```python
worlen = collections.OrderedDict(sorted(wordlen.items()))
worlen
```

```python
anagram_bylength = {}
print(worlen)

for length , words in worlen.items():
       anagram_bylength[length] =  {word: anagram_fast(word) for word in words if len(anagram_fast(words)) > 1}
```

```python
def parsefile(filename):
    return np.genfromtxt(filename,
                         delimiter = dly_delimiter,
                         usecols = dly_usecols,
                         dtype = dly_dtype,
                         names = dly_names)
```

```python
dly_delimiter = [11,4,2,4] + [5,1,1,1] * 31
dly_usecols = [1,2,3] + [4*i for i in range(1,32)]
dly_dtype = [np.int32,np.int32,(np.str_,4)] + [np.int32] * 31
dly_names = ['year','month','obs'] + [str(day) for day in range(1,31+1)]
```

```python
def unroll(record):
    startdate = np.datetime64('{}-{:02}'.format(record['year'],record['month']))
    dates = np.arange(startdate,startdate + np.timedelta64(1,'M'),np.timedelta64(1,'D'))
    
    rows = [(date,record[str(i+1)]/10) for i,date in enumerate(dates)]
    
    return np.array(rows,dtype=[('date','M8[D]'),('value','d')])
```

```python
def getobs(filename,obs):
    return np.concatenate([unroll(row) for row in parsefile(filename) if row[2] == obs])
```

```python
def fillnans(data):
    nan = np.isnan(data['value'])
    
    data['value'][nan] = np.interp(data['date'][nan],data['date'][~nan],data['value'][~nan])
```

```python
def selectyear(data,year):
    start = np.datetime64('{}'.format(year))
    end = start + np.timedelta64(1,'Y')
    
    return data[(data['date'] >= start) & (data['date'] < end)]['value']
```

```python
def topten(sex,year):
    simple = allyears_indexed.loc[sex,:,year].sort_values(by=['number'],ascending=False).reset_index()
    simple = simple.drop(['sex','year','number'],axis=1).head(10)
    
    simple.columns = [year]
    simple.index = simple.index + 1
    
    return simple
```

```python
topten('M',2009)
```

```python
def toptens(sex,year0,year1):
    years = [topten(sex,year) for year in range(year0,year1+1)]
    
    return years[0].join(years[1:])
```
